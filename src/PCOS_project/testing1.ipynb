```python
def score(row):
    protein_weight = 2
    fat_weight = -3
    fiber_weight = 2
    carb_weight = -1.5

    calc_score = (protein_weight * row['protein(g)'] + fat_weight * row['sugar(g)'] + fiber_weight * row['fiber(g)'] + carb_weight * row['carbs(g)'])
    return calc_score
```

```python
nutrition['PCOS_score'] = nutrition.apply(score, axis=1)
```

```python
def pcos_score(row):
    protein_weight = 2
    fat_weight = -3
    fiber_weight = 2
    carb_weight = -1.5

    calc_score = (protein_weight * row['Protein'] + fat_weight * row['Sat.Fat'] + fiber_weight * row['Fiber'] + carb_weight * row['Carbs'])
    return calc_score
```

```python
foods['PCOS_score'] = foods.apply(pcos_score, axis=1)
```

```python
def recommend(df, top=10, category=None):
    filtered = df.copy()

    if category:
        filtered = filtered[filtered['Category'] == category]

    recommendation = filtered.sort_values(by='PCOS_score', ascending=False)
    return recommendation.head(top)    
```

```python
recommendations = recommend(foods, 10, "Fruits A-F")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity

# Select features for similarity
features = foods[['Protein', 'Fat', 'Sat.Fat', 'Fiber', 'Carbs', 'PCOS_score']]

# Handle missing values
features = features.fillna(features.mean())

# Scale
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Cosine similarity
sim_matrix = cosine_similarity(features_scaled)
```

```python
def recommend_similar(food_name, top_n=5):
    if food_name not in foods['Food'].values:
        return f"⚠️ '{food_name}' not found in dataset. Try one of: {foods['Food'].sample(5).tolist()}"
    
    idx = foods[foods['Food'] == food_name].index[0]
    sim_scores = list(enumerate(sim_matrix[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    top_indices = [i for i, _ in sim_scores[1:top_n+1]]
    print(idx)
    return foods.iloc[top_indices][['Food', 'Category', 'PCOS_score']]

# Example
rec = recommend_similar("Shrimp", top_n=5)
```
